```python
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y')

```

```python
df_customizado = df[["categoria", "autoria", "titulo", "data", "link", "paragrafos","extra_01"]]
df_customizado
```

# Presidencia 2014

```python
presidencia_2014 = df_customizado[(df_customizado['data'].dt.year == 2014) & (df_customizado['categoria'].astype(str).str.contains('discursos')) & (df_customizado['autoria'].astype(str).str.contains('dilma-vana-rousseff-2011-2016')) ]
presidencia_2014

# luiz-inacio-lula-da-silva-2023-2026"
```

```python
# Coluna de interesse: presidente_2015['paragrafos']
coluna_paragrafos = presidencia_2014['paragrafos']

# 1. Converte a lista de parágrafos em uma única string por linha, separada por espaço.
# Usamos .apply(lambda x: ' '.join(x)) para unir os itens da lista em uma única string.
paragrafos_como_string = coluna_paragrafos.apply(lambda x: ' '.join(x))

# 2. Aplica a busca na nova Série de Strings
busca1 = presidencia_2014[paragrafos_como_string.str.contains("governança da internet", case=False, na=False)]
print (busca1)
# Exibe o resultado
print(f"Total de discursos encontrados: {len(busca1)}")
```
